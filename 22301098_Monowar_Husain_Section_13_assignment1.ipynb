```python
#QUESTION 01
from sympy import*
import numpy as np
x = symbols('x')
f = 9*x**3 - 7*x**2 + 3*x + 10
f = lambdify(x,f)
import numpy as np
p = np.arange(-5,7)
a = f(p)
print('f:',a)

print('GLOBAL (ABSOLUTE) MAXIMA;',max(a))
print('GLOBAL (ABSOLUTE) MINIMA:',min(a))
import matplotlib.pyplot as plt
plt.plot(p,a)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#QUESTION 02
from sympy import*
import numpy as np
x = symbols('x')
f = ln(x)
f = lambdify(x,f)
import numpy as np
p = np.arange(1,7)  #Negative values cannot be used in logarithmic functions.

a = f(p)
print('f:',a)

print('GLOBAL (ABSOLUTE) MAXIMA:',max(a))
print('GLOBAL (ABSOLUTE) MINIMA:',min(a))
import matplotlib.pyplot as plt
plt.plot(p,a)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
